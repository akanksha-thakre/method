## List Methods
### Appends()


```python
fruits =["apple", "banana", "cherry", "date"]
fruits.append("elderberry")
print(fruits)   
```

### Sort()

```python
def myFunc(e):
    return e['year']

cars = [
    {'car': 'Ford', 'year': 2005},          
    {'car': 'Mitsubishi', 'year': 2000},
    {'car': 'BMW', 'year': 2019},
    {'car': 'VW', 'year': 2011},
    {'car': 'Audi', 'year': 2005}
]
 
cars.sort(key=myFunc)
print(cars)
```

### Clear()


```python
car =['Ford', 'Mitsubishi', 'BMW', 'VW', 'Audi']
car.clear()
print(car)
```

### Copy()

```python
car =['Ford', 'Mitsubishi', 'BMW', 'VW', 'Audi']
car.copy()
print(car)
```

### Count()

```python
car =['Ford', 'Mitsubishi', 'BMW', 'VW', 'Audi']
x =car.count('BMW')
print(x)


fruits =["apple", "banana", "cherry", "date","cherry"]
x = fruits.count("cherry")
print(x)
```

### Extend()

```python
fruits =["apple", "banana", "cherry", "date"]
car =['Ford', 'Mitsubishi', 'BMW', 'VW', 'Audi']
fruits.extend(car)
print(fruits)
```

### Index()


```python
fruits =["apple", "banana", "cherry", "date","kiwi","mango","orange","papaya"]
x = fruits.index("mango",3)
print(x)
```

### Insert() 

```python
car =['Ford', 'Mitsubishi','VW', 'Audi']
car.insert(2,'BMW')
print(car)
```

### POP()

```python
fruits =["apple", "banana", "cherry", "date"]
fruits.insert(1, "grape")
print(fruits)   
```

### Remove()

```python
car =['Ford', 'Mitsubishi', 'BMW', 'VW', 'Audi']
car.remove('BMW')
print(car)
```

### Reverse()

```python
car = ['Ford', 'Mitsubishi', 'BMW', 'VW', 'Audi']
car.reverse()
print(car)
```
