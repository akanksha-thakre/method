## Strings Methods
### Capitalized


```python
### Capitalize()
txt = "enjoy !"
x = txt.capitalize()
print(x)
```

### Center()

```python
txt = "banana"
x = txt.center(20)
print(x)
```

### Casefold()

```python
txt = "hello, And Welcome to My World!"
x = txt.casefold()
print(x)
```

### Count()

```python

txt = "I love apples, apple are my favorite fruit"
x = txt.count("apple")
print(x)
```

### Index()

```python
txt = "Hello, welcome to my world"
x = txt.index("world")
print(x)
```

### Upper()

```python
txt = "Hello my friends , welcome to my world."
x = txt.upper()
print(x)
```

### Lower()

```python
txt ="Hello My Friends , Welcome To My World."
x =txt.lower()
print(x)
```

### Replace()

```python
txt ="Hello My Friends , Welcome To My World."
x = txt.replace("Friends", "Brothers")
print(x)
```

### isalnum

```python
txt ="company11"
x = txt.isalnum()
print(x)

txt ="company  21"
x = txt.isalpha()
print(x)
```

### Partition()

```python
txt = "I could eat bananas all day"
x = txt.partition("bananas")
print(x)
```

### Title()

```python
txt ="Welcome to my world"
x = txt.title()     
print(x)
```

### Strip()

```python
txt = " bananas "
x = txt.strip()
print("of all fruits", x, "is my favorite") 
```

### endswith()

```python
txt ="Hello, Welcome To My World."
x = txt.endswith(".")
print(x)


txt ="I eat all bananas"
x = txt.endswith("!")
print(x)
```

### Find()

```python
txt ="Hello, Welcome To My class."
x = txt.find("To")
print(x)
```

### join()

```python
myTuple = ("apple", "banana", "cherry")
x = "#".join(myTuple)   
print(x)
```
